```python
import numpy as np
```

```python
matrix_to_be_built = np.zeros((10,10), dtype=int)
matrix_to_be_built
```

```python
matrix_to_be_built.dtype
```

```python
for i in range(10):
    matrix_to_be_built[i][0] = i+1

for j in range(10):
    matrix_to_be_built[0][j] = j+1

matrix_to_be_built
```

```python
def build_matrix(i, j):
    if i == 0 or j == 0:
        return matrix_to_be_built[i][j]
    
    matrix_to_be_built[i][j-1] = build_matrix(i, j-1)
    matrix_to_be_built[i-1][j] = build_matrix(i-1, j)
    
    matrix_to_be_built[i][j] = matrix_to_be_built[i][j-1] + matrix_to_be_built[i-1][j]
    return matrix_to_be_built[i][j]

build_matrix(9,9)
matrix_to_be_built
```

```python
matrix_to_be_built.dtype
```

```python
print(matrix_to_be_built)
```

```python
bank = []

def snap_matrix(i, j):
    return matrix_to_be_built[i-2:i+1, j-2:j+1]
```

```python
snap_matrix(2,2)
```

```python
for i in range(10):
    if i < 2:
        continue
    for j in range(10):
        if j < 2:
            continue
        print(i, j)
        bank.append(snap_matrix(i, j))
```
